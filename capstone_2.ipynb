```python
import pandas as pd

# Assuming the datasets are stored in the following dataframes:
# orders_df, products_df, customers_df, stores_df, exchange_rates_df

# 1. Merge Orders with Customers on 'CustomerKey'
merged_data = pd.merge(sales, customers, on='CustomerKey', how='left')

# 2. Merge Orders-Customers with Stores on 'StoreKey'
merged_data = pd.merge(merged_data, stores, on='StoreKey', how='left')

# 3. Merge Orders-Customers-Stores with Products on 'ProductKey'
merged_data = pd.merge(merged_data, products, on='ProductKey', how='left')

# 4. Merge Orders-Customers-Stores-Products with Currency Exchange
# We need to match both the 'Currency Code' and 'Order Date' for an accurate exchange rate.
merged_data = pd.merge(merged_data, exchange_rates, 
                     left_on=['Currency Code', 'Order Date'], 
                     right_on=['Currency', 'Date'], 
                     how='left')

```

```python
# Drop the redundant columns after merge if needed
merged_data.drop(columns=['Currency', 'Date'], inplace=True)

# Now merged_df contains all the data merged from the five datasets.
```
